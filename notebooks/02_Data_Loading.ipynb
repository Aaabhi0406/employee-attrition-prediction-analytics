### Business Insight

- Employees with lower job satisfaction have a higher attrition rate.
- Attrition decreases as job satisfaction increases.
- HR should prioritize initiatives that improve employee satisfaction, such as career development, manager training, recognition programs, and work-life balance.
- This analysis identifies an association and does not establish a causal relationship.

```python
df_corr = df.copy()
```

```python
df_corr["Attrition"] = df_corr["Attrition"].map({
    "Yes": 1,
    "No": 0
})
```

```python
numerical_df = df_corr.select_dtypes(include=["int64", "float64"])
```

```python
correlation_matrix = numerical_df.corr()

correlation_matrix.head()
```

```python
plt.figure(figsize=(14, 10))

sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.show()
```

### Correlation Analysis

- Numerical features show relatively weak linear correlations with Attrition.
- Age, MonthlyIncome, JobLevel, and TotalWorkingYears have weak negative correlations with Attrition, suggesting employees with greater experience and higher salaries are somewhat less likely to leave.
- Strong positive correlations exist between JobLevel and MonthlyIncome (0.95), and between TotalWorkingYears and MonthlyIncome (0.77), reflecting expected career progression.
- Correlation analysis alone is insufficient for feature selection because important categorical variables, such as OverTime, are not fully represented by Pearson correlation.

```python
plt.figure(figsize=(12, 10))

sns.heatmap(
    correlation_matrix,
    cmap="coolwarm",
    center=0,
    square=True
)

plt.title("Correlation Matrix of Numerical Features")
plt.show()
```
