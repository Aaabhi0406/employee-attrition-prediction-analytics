# Business Problem

Employee attrition is one of the biggest challenges for organizations because losing experienced employees increases recruitment costs, training costs, and reduces productivity.

The objective of this project is to analyze employee data, identify the factors influencing attrition, and build a machine learning model capable of predicting whether an employee is likely to leave the company.

The insights generated from this project can help HR departments make informed decisions to improve employee retention.

# Business Objectives

- Understand employee characteristics.
- Identify factors affecting employee attrition.
- Perform exploratory data analysis.
- Build predictive machine learning models.
- Compare multiple models.
- Interpret business insights.
- Design an interactive Power BI dashboard.

# Machine Learning Problem

This is a Supervised Binary Classification problem.

Target Variable:

Attrition

Classes:

- Yes
- No
